## Exploratory Data Analysis

We calculate and draw plots based on frequency by players' performance

```python
%matplotlib inline
solo_KillDeathRatio = df['solo_KillDeathRatio'].tolist()
a = np.array(solo_KillDeathRatio)
hist, bins = np.histogram(a, [x / 10 for x in range(100)])
plt.hist(a, [x / 10 for x in range(100)] )
plt.title("histogram") 
plt.show()
```

### Histogram of solo player's kill-to-death ratio

```python
%matplotlib inline
duo_KillDeathRatio = df['duo_KillDeathRatio'].tolist()
a = np.array(duo_KillDeathRatio)
hist, bins = np.histogram(a, [x / 10 for x in range(100)])
plt.hist(a, [x / 10 for x in range(100)] )
plt.title("histogram") 
plt.show()
```

### Histogram of duo player's kill-to-death ratio

```python
%matplotlib inline
squad_KillDeathRatio = df['squad_KillDeathRatio'].tolist()
a = np.array(squad_KillDeathRatio)
hist, bins = np.histogram(a, [x / 10 for x in range(100)])
plt.hist(a, [x / 10 for x in range(100)] )
plt.title("histogram") 
plt.show()
```

### Histogram of squad player's kill-to-death ratio

All three columns are right skewed distributed<br>
Players tend to get 1 kdr regardless of how many teammates they have<br>

### Line chart

```python
%matplotlib inline
solo_KillDeathRatio = df['solo_KillDeathRatio'].tolist()
duo_KillDeathRatio = df['duo_KillDeathRatio'].tolist()
squad_KillDeathRatio = df['squad_KillDeathRatio'].tolist()

sa = np.array(solo_KillDeathRatio)
shist, sbins = np.histogram(sa, [x / 10 for x in range(100)])

da = np.array(duo_KillDeathRatio)
dhist, dbins = np.histogram(da, [x / 10 for x in range(100)])

sqa = np.array(squad_KillDeathRatio)
sqhist, sqbins = np.histogram(sqa, [x / 10 for x in range(100)])

plt.plot(range(99),shist,"b--",linewidth=1)
plt.plot(range(99),dhist,"r--",linewidth=1)
plt.plot(range(99),sqhist,"g--",linewidth=1)

plt.show()
```
